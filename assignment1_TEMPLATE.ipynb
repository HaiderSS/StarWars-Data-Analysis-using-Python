```python
# Loading Data
import pandas as pd
StarWarsDF = pd.read_csv("StarWars.csv", encoding = "unicode-escape", header=[0], skiprows=[1])
```

```python
# Renaming column according to my ease and protecting useful information
StarWarsDF = StarWarsDF.rename(columns = {
    'Which of the following Star Wars films have you seen? Please select all that apply.': 'Seen?(Yes/No) Episode I  The Phantom Menace',
    'Unnamed: 4': 'Seen?(Yes/No) Episode II  Attack of the Clones',
    'Unnamed: 5': 'Seen?(Yes/No) Episode III  Revenge of the Sith ',
    'Unnamed: 6': 'Seen?(Yes/No) Episode IV  A New Hope',
    'Unnamed: 7': 'Seen?(Yes/No) Episode V The Empire Strikes Back',
    'Unnamed: 8': 'Seen?(Yes/No) Episode VI Return of the Jedi', 
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'Rank(1-6) Episode I  The Phantom Menace',
    'Unnamed: 10': 'Rank(1-6) Episode II  Attack of the Clones',
    'Unnamed: 11': 'Rank(1-6) Episode III  Revenge of the Sith',
    'Unnamed: 12': 'Rank(1-6) Episode IV  A New Hope',
    'Unnamed: 13': 'Rank(1-6) Episode V The Empire Strikes Back',
    'Unnamed: 14': 'Rank(1-6) Episode VI Return of the Jedi', 
    'Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.': '(favorable/ unfavorable/ unfamiliar) Han Solo',
    'Unnamed: 16': '(favorable/ unfavorable/ unfamiliar) Luke Skywalker',
    'Unnamed: 17': '(favorable/ unfavorable/ unfamiliar) Princess Leia Organa',
    'Unnamed: 18': '(favorable/ unfavorable/ unfamiliar) Anakin Skywalker',
    'Unnamed: 19': '(favorable/ unfavorable/ unfamiliar) Obi Wan Kenobi',
    'Unnamed: 20': '(favorable/ unfavorable/ unfamiliar) Emperor Palpatine',
    'Unnamed: 21': '(favorable/ unfavorable/ unfamiliar) Darth Vader',
    'Unnamed: 22': '(favorable/ unfavorable/ unfamiliar) Lando Calrissian',
    'Unnamed: 23': '(favorable/ unfavorable/ unfamiliar) Boba Fett',
    'Unnamed: 24': '(favorable/ unfavorable/ unfamiliar) C-3P0',
    'Unnamed: 25': '(favorable/ unfavorable/ unfamiliar) R2 D2',
    'Unnamed: 26': '(favorable/ unfavorable/ unfamiliar) Jar Jar Binks',
    'Unnamed: 27': '(favorable/ unfavorable/ unfamiliar) Padme Amidala',
    'Unnamed: 28': '(favorable/ unfavorable/ unfamiliar) Yoda'
})
```

```python
# removing ResponderID as an unnessary column
StarWarsDF = StarWarsDF.iloc[:,1:]
```

```python
StarWarsDF.shape
```

```python
# to understanding anomalies in StarWarsDF (typos, whitespace etc)
for i in range(len(StarWarsDF.columns)):
    print(StarWarsDF.columns[i])
    print(StarWarsDF.iloc[:,i].value_counts())
    print()
```

```python
# Task-1.3 removing typos using mask
for i in 1,30,31,32: 
    mask1 = StarWarsDF[StarWarsDF.columns[i]] == "Yess"
    mask2 = StarWarsDF[StarWarsDF.columns[i]] == "Noo"
    mask3 = StarWarsDF[StarWarsDF.columns[i]] == "F"

    StarWarsDF[StarWarsDF.columns[i]].mask(mask1,"Yes",inplace=True)
    StarWarsDF[StarWarsDF.columns[i]].mask(mask2,"No",inplace=True)
    StarWarsDF[StarWarsDF.columns[i]].mask(mask3,"Female",inplace=True)    
```

```python
# Task- 1.4,1.5 converting all to upper case and removing whitespace except for float values
for i in range(0,7):
    StarWarsDF[StarWarsDF.columns[i]] = StarWarsDF[StarWarsDF.columns[i]].str.upper()
    StarWarsDF[StarWarsDF.columns[i]] = StarWarsDF[StarWarsDF.columns[i]].str.strip()
for i in range(14,37):
    StarWarsDF[StarWarsDF.columns[i]] = StarWarsDF[StarWarsDF.columns[i]].str.upper()
    StarWarsDF[StarWarsDF.columns[i]] = StarWarsDF[StarWarsDF.columns[i]].str.strip()
```

```python
# Task 1.6: Sanity checks
import numpy as np
for i in StarWarsDF.columns:
    print (StarWarsDF[i].value_counts().loc[ StarWarsDF[i].value_counts() < 2])
    
StarWarsDF['Age'] = StarWarsDF['Age'].replace({"500": np.nan})
```

```python
# Task-1.7 Handling missing values

# for ranking imputing their mean
StarWarsDF[['Rank(1-6) Episode I  The Phantom Menace',
           'Rank(1-6) Episode II  Attack of the Clones',
           'Rank(1-6) Episode III  Revenge of the Sith',
           'Rank(1-6) Episode IV  A New Hope',
           'Rank(1-6) Episode V The Empire Strikes Back',
           'Rank(1-6) Episode VI Return of the Jedi'
          ]] = StarWarsDF[['Rank(1-6) Episode I  The Phantom Menace',
           'Rank(1-6) Episode II  Attack of the Clones',
           'Rank(1-6) Episode III  Revenge of the Sith',
           'Rank(1-6) Episode IV  A New Hope',
           'Rank(1-6) Episode V The Empire Strikes Back',
           'Rank(1-6) Episode VI Return of the Jedi'
          ]].apply(lambda x: x.fillna(x.mean()),axis=0)

# for remaining object columns filling "NO RESPONSE"
for i in range(0,7):
    StarWarsDF[StarWarsDF.columns[i]] = StarWarsDF[StarWarsDF.columns[i]].fillna('NO RESPONSE')

for i in range(14,37):
    StarWarsDF[StarWarsDF.columns[i]] = StarWarsDF[StarWarsDF.columns[i]].fillna('NO RESPONSE')
```

```python
# Task-2.1 Rank star war films in order of preference
import matplotlib.pyplot as plt
ax = StarWarsDF[['Rank(1-6) Episode I  The Phantom Menace',
           'Rank(1-6) Episode II  Attack of the Clones',
           'Rank(1-6) Episode III  Revenge of the Sith',
           'Rank(1-6) Episode IV  A New Hope',
           'Rank(1-6) Episode V The Empire Strikes Back',
           'Rank(1-6) Episode VI Return of the Jedi'
          ]].astype(float).mean().sort_values(ascending = False)
```

```python
# plotting the values
ax.plot(kind='barh', fontsize=12, color = 'green')

for i, k in enumerate(ax):
    k = round(k,1)
    plt.text(k + 0.05, i, str(k), horizontalalignment='left')

plt.xlabel("Mean Ranking Score (Lesser the better)", fontsize=12)
plt.show()
```

```python
# Relation 2 : 
# Hypothesis : Star Trek Fans are also the fans of Star Wars Film franchise!
```

```python
# subsetting the dataframe according to our need by taking the nessary columns
df1 = StarWarsDF[["Do you consider yourself to be a fan of the Star Wars film franchise?",
                  "Do you consider yourself to be a fan of the Star Trek franchise?"]]
```

```python
# removing null values which we have renamed to "NO RESPONSE"
df1 = df1[df1['Do you consider yourself to be a fan of the Star Wars film franchise?'] != "NO RESPONSE"]
df1 =  df1[df1['Do you consider yourself to be a fan of the Star Trek franchise?'] != "NO RESPONSE"]
```

```python
# converting 'YES' to '1' and 'NO' to '0' so that we could apply the correlation function
df1 = df1.eq('YES').mul(1)
```

```python
df1.head()
```

```python
df1.corr()
```

```python
f = plt.figure(figsize=(10, 10))
plt.matshow(df1.corr(), fignum=f.number)
plt.xticks(range(df1.shape[1]), df1.columns, fontsize=6)
plt.yticks(range(df1.shape[1]), df1.columns, fontsize=8)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=14);
```

```python
# Result : Moderate correlation (0.49) between the two columns
```
